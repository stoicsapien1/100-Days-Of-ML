This is easy to implement technique

Used when we data missing completely at random and less than 5 % of the data is missing

There are two ways to impute things 

1. Mode imputation

2. Missing imputation: Simply add the "Missing" word in the empty places

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv",usecols=["GarageQual","FireplaceQu","SalePrice"])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df["GarageQual"].value_counts().sort_values(ascending=False).plot(kind="bar")
```

```python
df["GarageQual"].mode()
```

```python
df[df["GarageQual"]=="TA"]["SalePrice"].plot(kind="kde",label="TA")
df[df["GarageQual"].isnull()]["SalePrice"].plot(kind="kde",label="Missing")
plt.legend()
```

```python
temp=df[df["GarageQual"]=="TA"]["SalePrice"]
```

```python
df["GarageQual"].fillna("TA",inplace=True)
```

```python
df.isnull().sum()
```

```python
df["GarageQual"].value_counts().sort_values(ascending=False).plot(kind="bar")
```

```python
temp.plot(kind="kde")

df[df["GarageQual"]=="TA"]["SalePrice"].plot(kind="kde")
```

```python
df["FireplaceQu"].value_counts().sort_values(ascending=False).plot(kind="bar")
```

```python
df[df["FireplaceQu"]=="Gd"]["SalePrice"].plot(kind="kde",label="Gd")
df[df["FireplaceQu"].isnull()]["SalePrice"].plot(kind="kde",label="Missing")
plt.legend()
```

```python
temp=df[df["FireplaceQu"]=="Gd"]["SalePrice"]
```

```python
df["FireplaceQu"].fillna("Gd",inplace=True)
```

```python
df["FireplaceQu"].value_counts().plot(kind="bar")
```

```python
temp.plot(kind="kde")
df[df["FireplaceQu"]=="Gd"]["SalePrice"].plot(kind="kde")
```
