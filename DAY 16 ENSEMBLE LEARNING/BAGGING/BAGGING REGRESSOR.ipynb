```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=123)
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
knn=KNeighborsRegressor()
```

```python
lr.fit(X_train,y_train)
dt.fit(X_train,y_train)
knn.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
dt.score(X_test,y_test)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingRegressor
bag_reg=BaggingRegressor(random_state=1)
bag_reg.fit(X_train,y_train)
```

```python
bag_reg.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

# Define the parameter grid
params = {
    'estimator': [DecisionTreeRegressor(), LinearRegression(), KNeighborsRegressor()],
    'n_estimators': [20, 50, 100],
    'max_samples': [0.25, 0.5, 0.85,1.0],
    'max_features': [0.5, 1.0],
    'bootstrap': [True, False],
    'bootstrap_features': [True, False]
}

# Initialize the grid search
bag_reg_grid = GridSearchCV(BaggingRegressor(random_state=1), param_grid=params, cv=3, n_jobs=-1, verbose=1)
bag_reg_grid.fit(X_train, y_train)
```

```python
bag_reg_grid.best_params_
```

```python
bag_reg_grid.score(X_test,y_test)
```
