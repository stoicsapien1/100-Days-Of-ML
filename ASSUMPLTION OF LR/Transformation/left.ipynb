```python
from scipy.stats import weibull_min
import numpy as np

# Generate highly left-skewed data
data = weibull_min.rvs(c=0.5, scale=1, size=1000)
```

```python
import seaborn as sns
sns.histplot(data)
```

```python
data=data+1
```

```python
data.min()
```

```python
sns.histplot(data,kde=True)
```

```python
data_log=np.log(data)
```

```python
sns.histplot(data_log,kde=True)
```
