```python
from sklearn.datasets import make_regression
import pandas as pd

# Create a synthetic regression dataset
X, y = make_regression(n_samples=1000,        # Number of samples
                       n_features=10,         # Number of features
                       n_informative=8,       # Number of informative features
                       noise=0.2,             # Amount of noise in the data
                       random_state=42)       # For reproducibility

# Convert to a DataFrame
df = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(1, 11)])
df['Target'] = y

# Display the first few rows
print(df.head())
```

```python
df
```

```python
import seaborn as sns
sns.scatterplot(data=df,x="Feature_1",y="Target")
```

```python
df
```

```python
y=df["Target"]
```

```python
X=df.iloc[:,:-1]
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
```

```python
y_pred=lr.predict(X)
```

```python
residual=y-y_pred
```

```python
residual
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_pred,residual)
```

```python
residual
```

```python
data={"residual":residual}
```

```python
df1=pd.DataFrame(data)
```

```python
df1
```

```python
sns.histplot(df1["residual"],kde=True)
```

```python
## QQ plot of the residual to check normality assumption.
from scipy import stats

plt.figure(figsize=(8,6))
stats.probplot(df1["residual"],plot=plt)
plt.show()
```
