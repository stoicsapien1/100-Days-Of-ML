## Working with Dates

```python
date["date"]=pd.to_datetime(date["date"])
```

```python
date.head()
```

```python
date.info()
```

Extract year

```python
I have columnsdate["date_year"]=date["date"].dt.year
```

```python
date.head()
```

Extrct month

```python
date["date_month_no"]=date["date"].dt.month
```

```python
date.head()
```

```python
# Extract Month Name
date["date_month_name"]=date["date"].dt.month_name()
```

```python
date.head()
```

Extract days


```python
date["date_day"]=date["date"].dt.day
```

```python
date.head()
```

```python
#Day of the week
date["date_dow"]=date["date"].dt.dayofweek
```

```python
date.head()
```

```python
# Day of the week name
date["date_week_name"]=date["date"].dt.day_name()
```

```python
date.head()
```

```python
date.drop(columns=["product_id","city_id","orders"]).head()
```

```python
# is weekend?
date["date_is_weekend"]=np.where(date["date_week_name"].isin(["Saturday","Sunday"]),1,0)
```

```python
date.head()
```

Exract week of the year

```python
date["date_week"]=date["date"].dt.isocalendar().week
```

```python
date.head()
```

Extract Quater

```python
date["quarter"]=date["date"].dt.quarter
```

```python
date.head()
```

Extract Semester

```python
date["semester"]=np.where(date["quarter"].isin([1,2]),1,2)
```

```python
date.head()
```
