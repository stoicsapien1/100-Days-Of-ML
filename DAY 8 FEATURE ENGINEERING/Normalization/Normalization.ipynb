```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("wine_data.csv")
```

```python
df.head()
```

```python
df=df[["class_label","alcohol","malic_acid"]]
df
```

```python
color_dict={1:"red",2:"blue",3:"green"}
sns.scatterplot(df,x="alcohol",y="malic_acid",hue="class_label",palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop("class_label",axis=1),df["class_label"],test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```
