```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clas=DecisionTreeClassifier(max_depth=2)
```

```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
```

```python
X_train=iris.data
y_train=iris.target
```

```python
clas.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas)
```

## Classification

```python
from dtreeviz.trees import dtreeviz
viz=dtreeviz(clas,X_train,y_train,target_name="Price",feature_names=iris.feature_names,class_names=["setosa","versicolor","virginica"],scale=1.5)
```

```python
import dtreeviz

viz = dtreeviz.trees(clas, X_train, y_train, target_name="Price", feature_names=iris.feature_names, class_names=["setosa", "versicolor", "virginica"])
viz.view(scale=1.5)  # Uncomment to save with scale 1 (e.g., viz.save("decision_tree.svg"))
```
