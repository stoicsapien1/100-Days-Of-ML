```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
import seaborn as sns
```

```python
X=np.random.rand(100,1)-0.5
y=3*X[:,0]**2+0.05*np.random.randn(100)
```

```python
X
```

```python
import pandas as pd
df=pd.DataFrame()
df["X"]=X.reshape(100)
df["y"]=y
```

```python
df
```

```python
plt.scatter(df["X"],df["y"])
```

```python
df["pred1"]=df["y"].mean()
```

```python
df
```

```python
df["res1"]=df["y"]-df["pred1"]
```

```python
plt.scatter(df["X"],df["y"])
plt.plot(df["X"],df["pred1"],color="red")
```

```python
from sklearn.tree import DecisionTreeRegressor
tree1=DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree1.fit(df["X"].values.reshape(100,1),df["res1"].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree1)
plt.show()
```

```python
sns.kdeplot(np.linspace(-0.5,0.5,500))
```

```python
X_test=np.linspace(-0.5,0.5,500)
```

```python
y_pred=0.301479+tree1.predict(X_test.reshape(500,1))
```

```python
y_pred
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test,y_pred,linewidth=2,color="red")
plt.scatter(df["X"],df["y"])
```

```python
df["pred2"]=0.301479+tree1.predict(df["X"].values.reshape(100,1))
```

```python
df
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.scatter(df["X"],df["pred2"],linewidth=2,color="red")
plt.scatter(df["X"],df["y"])
```

```python
df["res2"]=df["y"]-df["pred2"]
```

```python
df
```

```python
tree2=DecisionTreeRegressor(max_leaf_nodes=8)
tree2.fit(df["X"].values.reshape(100,1),df["res2"].values)
```

```python
y_pred = 0.301479+sum(regressor.predict(X_test.reshape(-1,1)) for regressor in [tree1,tree2])
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test,y_pred,linewidth=2,color="red")
plt.scatter(df["X"],df["y"])
```
